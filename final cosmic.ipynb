```python
from itertools import combinations
import re

all_lists = [gnb5_sample, gnb4_sample, gnb3_sample, gnb1_sample, gnaq_sample, gnai2_sample, gnai1_sample, gna15_sample, gna14_sample, gna13_sample, gna11_sample, arrb2_sample, arr3_sample]

# Function to find common elements between multiple lists
def find_common_elements(*lists):
    # Find intersection using set intersection
    common_elements = set(lists[0]).intersection(*lists[1:])
    return list(common_elements)

# Function to get variable name from its value
def get_var_name(var, namespace):
    return [name for name, value in namespace.items() if value is var]

# Open a text file for writing
with open('cosmic_common_elements.txt', 'w') as file:
    # Find common elements for different combinations (2 by 2, 3 by 3)
    for r in range(2, len(all_lists) + 1):
        file.write(f"Checking {r} by {r} combinations:\n")
        for combination in combinations(all_lists, r):
            # Get variable names for printing
            list_names = [get_var_name(var, globals())[0] for var in combination]
            # Remove invalid characters from variable names
            list_names = [re.sub(r'\W+', '', name) for name in list_names]
            cosmic_common_elements = find_common_elements(*combination)
            if cosmic_common_elements:
                file.write(f"Common elements in ({', '.join(list_names)}): {cosmic_common_elements}\n")

print("Results written to 'common_elements.txt'")
```
